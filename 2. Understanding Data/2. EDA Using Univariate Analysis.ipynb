### Whenever we receive data, our starting point involves asking seven fundamental questions to establish a foundational understanding or basic understanding. While there's no rigid rule, this method helps guide our understanding.

### The questions are: 
    1. How big is the data?
    2. How does the data look like?
    3. What is datatype of columns?
    4. Are there any missing values?
    5. How does the data look like mathematically?
    6. Are there any duplicate values?
    7. How is the correlation between cols?

### Once we have a basic understanding of the data, we move on to Exploratory Data Analysis (EDA). 

### EDA is like investigating a treasure map to find clues and understand what's there. It involves diving deeper into the data to discover patterns, relationships, and interesting insights.

### Methods used in EDA include:
    1. Univariate Analysis
    2. Bivariate Analysis
    3. Multivariate Analysis
    4. Summary Statistics
    5. Data Visualization
    6. Outlier Detection
    7. Missing Values Handling etc.

## Univariate Analysis: 
###### This examines one variable (one column) at a time to understand its distribution and characteristics. For example, looking at histograms to see how values are spread.

###### Usually data is of two types mainly:
    1. Numerical data
    2. Categorical Data

while performing univariate analysis, we need to first question ourselevs about whether the data type in column is numerical or categorical and depending on that we perform plotting accrodingly.

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

# 1. Categorical Data

## a. Countplot

###### count plot helps us understand the frequency or count of each category in that particular column

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
df['Embarked'].value_counts().plot(kind='bar')
```

```python
# Plot count of 'Embarked' values using seaborn
sns.countplot(data=df, x='Embarked')
#sns.countplot(df['Embarked'])
#df['Embarked'].value_counts().plot(kind='bar')
```

```python
sns.countplot(data=df, x='Survived')
```

## b. PieChart

###### Pie chart can be used to know the distribution percentage of each category in that specific column.

```python
#using matplot
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
#using matplot
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
#using matplot
df['Embarked'].value_counts().plot(kind='pie',autopct='%.2f')
```

# 2. Numerical Data


###### When we have numerical data we can explore in various ways. one way is to convert that numerical data to categorical data. suppose Here Age is numerical data but we can see the distribution of in ranges like 0-10 as one class, 10-20 as next class and so on.

## a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'])
```

```python
#need mention bins to alter it to our use
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=8)
```

```python
#need mention bins to alter it to our use
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=5)
```

## b. Distplot

###### Gives probability density function (pdf) graph

```python
sns.distplot(df['Age'])
```

```python
sns.displot(df['Age'])
```

```python
# when you want kde with histrogram
sns.displot(df['Age'], kde=True)
```

```python
# when you want only kde without histrogram
sns.displot(df['Age'], kind='kde')
```

## c. Boxplot

###### Gives 5 point summary:
    1. Minimum (Q1-1.5*IQR)
    2. Q1 (25th Percentile)
    3. Median
    4. Q3 (75th Percentile)
    5. Maximuum (Q3+1.5*IQR)
    
    Here, Interquartile Range (IQR)=Q3-Q1
    
#### Note: Anything That lies beyond the range of [minimum, maximum] is considered as outliers.


which means all those values which are less than Minimum or greater than Maximum are outliers


```python
sns.boxplot(df['Age'])
```

```python
sns.boxplot(df['Fare'])
```
