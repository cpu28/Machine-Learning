### Whenever we receive data, our starting point involves asking seven fundamental questions to establish a foundational understanding or basic understanding. While there's no rigid rule, this method helps guide our understanding.

### The questions are: 
    1. How big is the data?
    2. How does the data look like?
    3. What is datatype of columns?
    4. Are there any missing values?
    5. How does the data look like mathematically?
    6. Are there any duplicate values?
    7. How is the correlation between cols?

### Once we have a basic understanding of the data, we move on to Exploratory Data Analysis (EDA). 

### EDA is like investigating a treasure map to find clues and understand what's there. It involves diving deeper into the data to discover patterns, relationships, and interesting insights.

### Methods used in EDA include:
    1. Univariate Analysis
    2. Bivariate Analysis
    3. Multivariate Analysis
    4. Summary Statistics
    5. Data Visualization
    6. Outlier Detection
    7. Missing Values Handling etc.

## Univariate Analysis: 
    This examines one variable at a time to understand its distribution and characteristics. For example, looking at histograms to see how values are spread.

### After Univariate Analysis, can proceed to Bivariate and MultiVariate Analysis

## Bivariate Analysis: 
    This analyzes the relationship between two variables. For instance, scatter plots can show how two variables relate to each other.

## Multivariate Analysis: 
    This explores relationships between multiple variables at once. Techniques like heatmaps or correlation matrices help understand interactions among several variables.

## 1. Scatterplot (Numerical - Numerical)

```python
#Bivariate
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
#Multivariate (3 variables here)
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
#Multivariate (4 variables here)
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker')
```

##### Important

```python
#Multivariate (5 variables here)
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker', size='size')
```

###### Why to Use plot.show(): 
When you create a plot in Matplotlib or Seaborn without using plt.show() in a Jupyter Notebook or IPython environment, the plot will typically display automatically. However, if you're working in a script or an environment that doesn't automatically render plots, such as a Python script or some Python integrated development environments (IDEs), the plot won't appear until you explicitly call plt.show().

In Jupyter Notebooks, using plt.show() is not always necessary to display plots. Plots will usually render automatically after the code cell in which they were created. However, when you have multiple plots or want to explicitly display the plot at a specific point, it's good practice to use plt.show() to ensure that the plot appears.

It's worth noting that in interactive mode or some specialized environments, not using plt.show() might still display the plot, but it's a good habit to include plt.show() to guarantee that your plots are displayed correctly regardless of the environment.

```python
#Multivariate (5 variables here)
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker', size='size')
plt.show()
```

## 4. Distplot (Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

```python
#using displot
sns.displot(titanic[titanic['Survived'] == 0]['Age'], kind='kde', fill=False)
sns.displot(titanic[titanic['Survived'] == 1]['Age'], kind='kde', fill=False)
```

```python
#using kdeplot
sns.kdeplot(titanic[titanic['Survived'] == 0]['Age'], fill=False)
sns.kdeplot(titanic[titanic['Survived'] == 1]['Age'], fill=False)
```

```python
#Bi Variate using kdeplot, same as above just chaging parameters format
sns.kdeplot(x='Age', data=titanic[titanic['Survived'] == 0],fill=False)
sns.kdeplot(x='Age', data=titanic[titanic['Survived'] == 1],fill=False)
```

```python
#Bi Variate using kdeplot, same as above just chaging parameters format
sns.kdeplot(x='Age', data=titanic, hue='Survived',fill=False)
```

##### Why above two are diffrent? eventhough logicallly, they should be same???

```python
sns.kdeplot(x='Age', data=titanic[titanic['Survived'] == 0],hue='Sex',palette='deep',fill=False)
```

```python
sns.kdeplot(x='Age', data=titanic[titanic['Survived'] == 1],hue='Sex',palette='pastel',fill=False)
```
